## Question 1: 
## “Which state has the highest number of colleges and universities per capita?”

```python
df1 = university[["STATE","POPULATION"]]
df1
```

```python
df1 = df1.replace(',','', regex=True)
df1["POPULATION"] = pd.to_numeric(df1["POPULATION"])
df1
```

```python
x = df1["STATE"].value_counts()
x=x.to_list()
len(x)
```

```python
y = df1.drop_duplicates().reset_index().drop("index",axis=1)
y["COUNT"] = x
y["COLLEGES PER CAPITA"] = y["COUNT"]/y["POPULATION"]
y[y['COLLEGES PER CAPITA']==y["COLLEGES PER CAPITA"].max()]
```

```python
fig, ax = plt.subplots(figsize=(30,10))
#ax.scatter(y["STATE"],y["UNI_PCP"])
ax.bar(y["STATE"],y["COLLEGES PER CAPITA"],color = "orange",edgecolor="blue")
#ax.plot(y["STATE"],y["UNI_PCP"])
ax.set_xlabel('State', fontsize=20)
ax.set_ylabel('College Per Capita', fontsize=20)
```
